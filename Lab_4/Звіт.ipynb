```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

My_graph = nx.Graph()
My_graph = nx.read_gexf('graph_file.gexf')
nodes_positions = {'A': [1, 4], 'C': [7, 4], 'B': [4, 10], 'D': [4, 7], 'E': [4, 5], 'J': [11, 8], 'G': [4.5, 7.5],
                   'F': [3.5, 7.5], 'I': [10, 6], 'H': [9, 9]}
nodes = nodes_positions.keys()
nx.draw(My_graph, pos=nodes_positions, with_labels=True, font_weight='bold', font_color='white')
plt.show()
plt.savefig('graph_1.jpeg')
```

Зображення графу

```python
for i in nx.connected_components(My_graph):
    print("----------------------")
    print(f'The number of edges is {My_graph.subgraph(i).number_of_edges()}')
    print(f'The number of nodes is {My_graph.subgraph(i).number_of_nodes()}')
    for node in My_graph.subgraph(i):
        print(f'The degree of {node} is {My_graph.subgraph(i).degree({node})}')
    print('Eccentricity: ', nx.eccentricity(My_graph.subgraph(i).subgraph(i)))
    print('Radius is', nx.radius(My_graph.subgraph(i)))
    print('Diameter is', nx.diameter(My_graph.subgraph(i)))
```

Кількість вершин, ребер, степені вершин, ексцентриситету, радіуса та діаметра по кожній компоненті зв'язності

```python
edges_diameter_list = []


def diameter():
    for v in p:
        for w in p:
            if nx.shortest_path_length(My_graph.subgraph(p), v, w) == nx.diameter(My_graph.subgraph(p)):
                shortest_path = nx.shortest_path(My_graph.subgraph(p), v, w)
                edges_diameter = [(a, b) for a, b in zip(shortest_path, shortest_path[1:])]
                edges_diameter_list.extend(edges_diameter)
                return


for p in nx.connected_components(My_graph):
    if nx.number_of_nodes(My_graph.subgraph(p)) > 1:
        diameter()
print(edges_diameter_list)
nx.draw(My_graph, pos=nodes_positions, with_labels=True, font_weight='bold', font_color='white')
nx.draw_networkx_edges(My_graph, pos=nodes_positions, edgelist=edges_diameter_list, edge_color='r')
plt.show()
plt.savefig('graph_2.jpeg')
```

Зображення будь-якого діаметру по кожній компоненті зв'язності

```python
edges_dfs_list = []


for z in nx.connected_components(My_graph):
    for q, y in nx.dfs_edges(My_graph.subgraph(z)):
        edges_dfs = [(a, b) for a, b in nx.dfs_edges(My_graph.subgraph(z))]
        edges_dfs_list.extend(edges_dfs)
nx.draw(My_graph, pos=nodes_positions, with_labels=True, font_weight='bold', font_color='white')
nx.draw_networkx_edges(My_graph, pos=nodes_positions, edgelist=edges_dfs_list, edge_color='y')
plt.show()
plt.savefig('graph_3.jpeg')
```
